```python
from pathlib import Path
```

```python
fait = []
for csv_file in liste_csv:
    if Path("D:/Repositories/Data Italy/Italie/Italie/"+csv_file).stat().st_size > 10000:
        df = pd.read_csv("D:/Repositories/Data Italy/Italie/Italie/" + csv_file, sep=";")
        df.columns = ['Codice_sensore', 'year', 'month', 'day', 'hour', 'minutes', 'Precipitazione', 'Neve', 'Codice_stazione']
        date = pd.to_datetime(df.iloc[:,1:6])
        df2 = df.drop(['year', 'month', 'day', 'hour', 'minutes'], axis=1)
        df2.insert(0, "date", date)
        df2 = df2.loc[df2.Precipitazione != "  Dato mancante"]
        df2['Precipitazione'] = df2['Precipitazione'].astype(float)
        df2['Neve'] = df2['Neve'].astype(float)
        df2 = df2[~df2.Neve.isna()]
        df2 = df2[df2.Precipitazione >= 0]
        df2 = df2[df2.Precipitazione < 110] #valeur arbitraire
        try:
            write_api.write(
                record=df2,
                bucket="rainfall",
                data_frame_measurement_name="rainfall",
                data_frame_tag_columns=["Codice_sensore", "Codice_stazione"],
                data_frame_timestamp_column="date",
            )
            fait.append(csv_file)
        except AttributeError as e:
            print(e)
    print(csv_file)
```

```python
import geopandas as gpd

dfita = gpd.read_file('D:/Repositories/Data Italy/Data Italy/stations.gpkg') #besoin de stations.gpkg
dfita["rain"] = np.nan
query_api = client.query_api()
for id in dfita.id:
    query = f"""from(bucket: "rainfall")
     |> range(start: 2014-03-26T07:00:00Z, stop: 2014-03-27T23:00:01Z)
     |> filter(fn: (r) => r._measurement == "rainfall")
     |> filter(fn: (r) => r._field == "Precipitazione")
     |> filter(fn: (r) => r.Codice_sensore == "{id}")
     |> sum()"""
    tables = query_api.query(query, org="IFPEN")
    for table in tables:
        for record in table.records:
            val = record.get_value()
            idx = dfita.loc[dfita.id == id].index
            dfita.iloc[idx, -1] = val
dfita
```
